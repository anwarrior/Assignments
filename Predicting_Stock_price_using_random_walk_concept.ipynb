```python
#import Libraries
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

#Coin Toss
toss = random.uniform(0,1)

Price = 158
j = 0
index_up = 1.01       #Growth Rate
index_down = 0.99     #Rate Of decline
y_axis = []              #Represent stock price of 100 days
x_axis = range(0,100)    #Represents 100 days

while j <= 99:
  y_axis.append(Price)
  toss = random.uniform(0,1)
  if toss > 0.5:              #Represents Heads
    Price = Price*index_up
  elif toss < 0.5:            #Represents Tails
    Price = Price*index_down
  j = j+1

x_axis = range(0,100)         #Generating Sequence from 0 to 100
plt.plot(x_axis , y_axis,'sb-',  markerfacecolor='none')
plt.yticks(fontsize = 16)
plt.xticks(np.arange(0,100+1, 10), fontsize = 16)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)    #Altering size of chart
plt.xlabel('Days', fontsize=26)
plt.ylabel('Price (in $)', fontsize=26)
plt.title('Predicted Boeing Company Stock Price' , fontsize = 25)
fig.savefig('test2png.png', dpi=100)         #saving graph as img in png format
plt.show()
```
